# barplot seaborn 


```python
import seaborn as sns
import matplotlib.pyplot as plt 
import scipy 
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day, y=tips_df.total_bill,hue=tips_df.sex)
```

```python
sns.barplot(x='day', y='total_bill',hue='sex',data=tips_df) 
# alternating way of writing aboce coad
```

```python
order = ['Sun','Thur','Fri','Sat']
sns.barplot(x='day', y='total_bill',hue='sex',data=tips_df,order = order) 
```

```python
hue_order=["Male","Female"]

order = ['Sun','Thur','Fri','Sat']
sns.barplot(x='day', y='total_bill',hue='sex',data=tips_df,hue_order = hue_order) 
```

```python
order = ['Sun','Thur','Fri','Sat']
sns.barplot(x='day', y='total_bill',hue='sex',data=tips_df,order = order,ci=12)
# ci stand for confidance interval and this is that black line over the graph !
```

```python
sns.barplot(x='day', y='total_bill',hue='sex',data=tips_df,hue_order = hue_order,capsize=0.3) 
```

```python
sns.barplot(x='day', y='total_bill',hue='sex',data=tips_df,hue_order = hue_order,dodge=False) 
```
