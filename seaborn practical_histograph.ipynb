# histograph in seaborn using distplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats 
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.distplot(tips_df['size'])
```

```python
sns.distplot(tips_df['tip'])
```

```python
sns.distplot(tips_df['total_bill'])
```

```python
sns.distplot(tips_df['size'],bins=10)
```

```python
sns.distplot(tips_df['total_bill'],bins= 244) # this line is called as kernal density extiater whihc is special function in displot or dostplot in seaborn
```

```python
sns.distplot(tips_df['total_bill'],color='r',vertical=True)
```

```python
sns.distplot(tips_df['total_bill'],color='r',axlabel="Total Bill")
```

```python
sns.distplot(tips_df['total_bill'],color='r',label="total bill")
plt.legend()
plt.title("histogram of Totalbill",fontsize=20)
```
