# Advance Histograph

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
plt.figure(figsize=(14,9))
sns.set()
sns.distplot(tips_df["total_bill"],color="r")
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(14,9))
sns.set()
plt.xticks(bins)
sns.distplot(tips_df["total_bill"],color="r",bins=bins)
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(14,9))
sns.set()
plt.xticks(bins)
sns.distplot(tips_df["total_bill"],color="r",bins=bins,
            hist_kws={'color':'m','edgecolor':'r','linewidth':5,'alpha':0.9})
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(14,9))
sns.set()
plt.xticks(bins)
sns.distplot(tips_df["total_bill"],color="r",bins=bins,
            hist_kws={'color':'m','edgecolor':'r','linewidth':5,'alpha':0.9},
            
             kde_kws={'color':'r','linewidth':5,'alpha':0.9}
            )
```
